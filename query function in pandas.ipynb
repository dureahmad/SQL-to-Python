```python
import pandas as pd
import numpy as np
# import scipy

df =pd.read_csv('cars.csv') 
```

```python
df.query('gear == "5"').sum()
```

```python
mpg = df.mpg
mpg_mean= mpg.mean()
mpg_min = mpg.min()
mpg_max = mpg.max()
print('Mean value of Miles per Gallon is...', mpg_mean)
print('Minimum value of Miles per Gallon is...', mpg_min)
print('Maximum value of Miles per Gallon is...', mpg_max)


```
