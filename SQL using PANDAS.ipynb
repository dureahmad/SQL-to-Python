## JOIN Clause
Join clasuse is used to combine rows from two or more tables, baseed on a related columns between them

**Join airports and airport_freq table where ident is "KLAX' and display the ident,description and frequency_mhz**

```python
airport_freq = pd.read_csv('airport-frequencies.csv')
airport_freq.head()
```

```python
ps.sqldf('select airport_ident,airport_freq.type, description,frequency_mhz from\
          airport_freq join airports on airport_freq.airport_ref = airports.id\
          where airports.ident="KLZU"')
```

**In Panadas we use**

```python
airport_freq.merge(airports[airports.ident=='KLZU']['id'], left_on='airport_ref',right_on = 'id',
                   how = 'inner')[['airport_ident','type','description','frequency_mhz']]
```

## SORTING
**To sort the results in ascending or descending order we will use "Order By" clause** 


```python
ps.sqldf('select * from airport_freq where airport_ident = "KLZU" order by type')
```

```python
# Now if we want to display in descending order
ps.sqldf('select * from airport_freq where airport_ident = "KLZU" order by type desc')
```

**Using pandas**

```python
airport_freq[airport_freq.airport_ident == 'KLZU'].sort_values('type', ascending =False)
```

### Group By Clause
**This clause is used with nther aggregate functions such as "sum, avg, count, min, max" to group mthe result-set by one or more columns**

**Counting airports groups by country**

```python
ps.sqldf('select iso_country, count(*) from airports group by iso_country')
```

**In Pandas**

```python
airports.groupby(['iso_country']).size()
```

**Count the airports grouped and order by country anf type using ascending order-  (Using SQL)**

```python
ps.sqldf('select iso_country,type, count(*) from airports group by iso_country,\
          type order by iso_country,type')
```

**Using Pandas**

```python
airports.groupby(['iso_country','type']).size()
```

**Counting the number of airports grouped by country and type and placed in descending order (Using SQL)**

```python
ps.sqldf('select iso_country,type,count(*) as size from airports\
          group by iso_country, type order by iso_country,count(*) desc')
```

**Using Pandas**


```python
airports.groupby(['iso_country','type']).size().to_frame('size').\
                sort_values(['iso_country','size'],ascending=[True,False])
```

**Counting Uniqe values -(using SQL)**

```python
ps.sqldf('select type, count(*) from airports group by type')
```

```python
# Using Pandas
airports.type.value_counts()
```

### Having Clause

**This clasue is used with aggregate function. It is used to filter the grouped data using 'HAVING' condition**

```python
ps.sqldf('select type,count(*) from airports where iso_country ="US"\
 group by type having count(*) > 2000 order by count(*) desc')
```

**Using pandas (we will use filter() alongwith lambda function**

```python
airports[airports.iso_country =='US'].groupby('type').filter(lambda x: len(x) >1000).\
         groupby('type').size().sort_values(ascending=False)
```
